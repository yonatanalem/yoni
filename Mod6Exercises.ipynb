```python
import pandas as pd
import numpy as np

location = 'datasets/tamiami.csv'
df = pd.read_csv(location)
```

```python
df.rename(columns={'Cart Location':'Location', 'Hot Dog Sales': 'Sales', 'Employees in Nearby Office Buildings': 'Employees', 'Num of Nearby Restaurants':'Restaurants', 'Num of Other Food Carts Nearby': 'Foodcarts'},inplace=True)
df.head()
```

```python
df.corr()
```

### Exercise 2:

Using the dataframe from the previous exercise, choose features (columns) to create a linear regression formula to predict sales. Try it with and without the y-intercept. How does it make a difference? Does adding or removing features in your model formula make a difference in the output?

```python
import statsmodels.formula.api as smf
result = smf.ols('Sales ~ Employees + Foodcarts + Price', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as smf
result = smf.ols('Sales ~ Employees + Foodcarts + Price - 1', data=df).fit()
result.summary()
```
